```python
import numpy as np
from scipy.special import iv  # 修正贝塞尔函数
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 定义参数
G0 = 0.2
c_B = 10.0
lambda_val = 0.5

# 定义ODE方程
def dRdt(t, R):
    x = np.sqrt(lambda_val) * R
    I0 = iv(0, x)  # 第一类修正贝塞尔函数0阶
    I1 = iv(1, x)  # 第一类修正贝塞尔函数1阶
    return (G0 * c_B * I1) / (np.sqrt(lambda_val) * I0)

# 初始条件和时间范围
R0 = [0.5]  # 初始半径值（可修改）
t_span = (0, 0.2)  # 时间范围（秒）
t_eval = np.linspace(*t_span, 21)  # 评估点

# 求解ODE
sol = solve_ivp(dRdt, t_span, R0, t_eval=t_eval, method='RK45')
```

```python
numerical_r = [0.5, 0.50492879, 0.50990373, 0.51492714, 0.51999802, 0.52511658, 0.53028530, 0.53550129, 0.54076960, 0.54608502, 0.55145267, 0.55686895, 0.56233792, 0.56785786, 0.57342922, 0.57905432, 0.58473201, 0.59046307, 0.59624741, 0.60208747, 0.60798085]
```

```python
plt.figure(dpi=600)
plt.plot(sol.t, sol.y.flatten(), 'b--', alpha = 0.7, linewidth=2, label='Analytical R(t)')  
plt.plot(sol.t, numerical_r, 'ro-', markersize=3, alpha = 0.7, linewidth=1, label='Numerical R(t)')  
plt.title('Radius Evolution')
plt.xlabel('Time t')
plt.ylabel('Radius R(t)')
plt.grid(True)
plt.legend()
plt.xticks(fontsize=8) 
plt.yticks(fontsize=8) 
plt.show()
```
