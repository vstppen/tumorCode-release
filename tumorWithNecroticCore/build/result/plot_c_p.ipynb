```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

def plot_field_evolution(indices, delta_t, extent=[-5.0, 5.0, -5.0, 5.0]):
    num_plots = len(indices)
    fig, axes = plt.subplots(2, num_plots, figsize=(4.5 * num_plots, 8), dpi=600)
    
    c_data_list = []
    p_data_list = []
    
    # Load data for all requested indices
    for idx in indices:
        c_data_list.append(np.loadtxt(f'./solutions/c_{idx}.txt'))
        p_data_list.append(np.loadtxt(f'./solutions/p_{idx}.txt'))
        
    # Determine global min and max for shared colormaps
    c_vmin, c_vmax = np.min(c_data_list), np.max(c_data_list)
    p_vmin, p_vmax = np.min(p_data_list), np.max(p_data_list)
    
    for i, idx in enumerate(indices):
        current_time = idx * delta_t
        
        # Top row: Nutrient Concentration
        ax_c = axes[0, i]
        im_c = ax_c.imshow(c_data_list[i], extent=extent, origin='lower', 
                           cmap='viridis', vmin=c_vmin, vmax=c_vmax)
        ax_c.set_title(f'$c(\\cdot, t={current_time:.2f})$')
        ax_c.set_xlabel('$x$')
        ax_c.set_ylabel('$y$')
        
        if i == num_plots - 1:
            divider = make_axes_locatable(ax_c)
            cax = divider.append_axes("right", size="5%", pad=0.1)
            fig.colorbar(im_c, cax=cax)

        # Bottom row: Pressure
        ax_p = axes[1, i]
        im_p = ax_p.imshow(p_data_list[i], extent=extent, origin='lower', 
                           cmap='plasma', vmin=p_vmin, vmax=p_vmax)
        ax_p.set_title(f'$p(\\cdot, t={current_time:.2f})$')
        ax_p.set_xlabel('$x$')
        ax_p.set_ylabel('$y$')
        
        if i == num_plots - 1:
            divider = make_axes_locatable(ax_p)
            cax = divider.append_axes("right", size="5%", pad=0.1)
            fig.colorbar(im_p, cax=cax)

    plt.tight_layout()
    plt.show()
```

```python
plot_field_evolution([10, 20, 30, 40], 0.01)
```
